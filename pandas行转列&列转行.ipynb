```python
df3 = pd.DataFrame(data = {"姓名":['C罗','坤坤','梅西']
                    ,"年龄":[10,20,19],
                    "特长":['速度,力量,弹跳','唱,跳,rap,篮球','灵巧,速度,盘带']},columns = ['姓名','年龄','特长'])
```

### 1.使用原生的stack和unstack来实现

```python
#  因为是要对特长这个字段进行分解 
#  所以将除了“特长”之外的其他字段设置为索引
df3 = df3.set_index(['姓名','年龄'])
```

```python
# 筛选出“爱好”这个字段，此时得到的是一个具有二级索引的series
s = df3['特长']
```

```python
# 对这个series进行分解 ，得到了一个具有二级索引的dataframe
df3 = s.str.split(",",expand = True)
```

```python
# 调用stack，将df3编程一个三级索引的series
s = df3.stack()
```

```python
# 下面直接reset_index并且删除没有用的0,1,2就好了
s.reset_index(drop = True,level=-1).reset_index().rename(columns = {0:"特长"})
```

```python
# 用一行代码写完就是：
df3.set_index(['姓名','年龄'])['特长'].str.split(",",expand = True).stack().reset_index(drop = True,level = -1).reset_index().rename(columns = {0:"特长"})
```

```python
"""
练习4 进行如下转换
id                                       info
0  001                     {'name': 'mashiro', 'age': 17, 'gender': 'female'}
1  002                      {'name': 'satori', 'age': 16, 'gender': 'female'}
2  003  {'name': 'nagisa', 'age': 21, 'gender': 'female', 'where': 'clannad'}

id     name  age  gender    where
0  001  mashiro   17  female      NaN
1  002   satori   16  female      NaN
2  003   nagisa   21  female  clannad
"""
```

```python
df4 = pd.DataFrame(columns=['id','info'],data = {'id':['001','002','003'],'info':[{'name': 'mashiro', 'age': 17, 'gender': 'female'},
                                                     {'name': 'satori', 'age': 16, 'gender': 'female'},
                 
                                                                                  {'name': 'nagisa', 'age': 21, 'gender': 'female', 'where': 'clannad'}]})
```

```python
# 这个比较简单，只需要对info这一列使用apply(pd.Series)，就可以将字典变为多个字段，key为列名
df4['info'].apply(pd.Series).join(df4['id'])
```

```python
# 如果传进来的列表嵌套的字典被用""修饰的话，刚刚那个方法行不通，需要首先使用eval函数将str转换成字典，再对字典做apply操作
df4_str = pd.DataFrame(columns=['id','info'],data = {'id':['001','002','003'],'info':["{'name': 'mashiro', 'age': 17, 'gender': 'female'}",
                                                     "{'name': 'satori', 'age': 16, 'gender': 'female'}",
                                                     "{'name': 'nagisa', 'age': 21, 'gender': 'female', 'where': 'clannad'}"]})
```

```python
df4_str['info'] = df4_str['info'].apply(lambda x:dict(eval(x)))
```

```python
df4_str['info'].apply(pd.Series).join(df4_str['id'])
```
